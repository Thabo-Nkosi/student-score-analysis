```python
import pandas as pd
import matplotlib.pyplot as plt

students= {
    "Name": ["Thabo", "Lerato", "Sipho", "Noma"],
    "Subject": ["python", "python", "python", "python"],
    "Score": [88, 65, 45, 74]
}
df = pd.DataFrame(students)
print(df)

#find who passed
print("--------------------------------------------")
df['passResults'] = df['Score'].apply(lambda score: "pass" if score >=50 else "fail")
print(df)

#Finding the average score
print("\n------------------------------------------")
df['average'] = df['Score'].mean()
print(df)

print("\n----------------------------------------------------------------------")
def fund(Score):
  if Score >= 75:
    return 'Pass with a distiction'

  elif Score < 75 and Score > 50:
    return 'Normal pass'
  else:
    return 'You need to work hard'

df['result'] = df['Score'].apply(fund)
print(df)

# Define colors based on pass/fail
colors = df['passResults'].map({'pass': 'green', 'fail': 'red'})

#creating a bar chart
plt.figure(figsize=(8,4))
bars =plt.bar(df["Name"], df["Score"], color =colors)

# Add score labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1, f'{yval}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Add title and labels
plt.title("Thabo's Student Score Report", fontsize=14, fontweight='bold')
plt.xlabel("Student Name")
plt.ylabel("Score")
# Set Y-axis limit for better spacing
plt.ylim(0, 100)
plt.legend()

# Show horizontal line for average
plt.axhline(y=df["Score"].mean(), color='red', linestyle='--', label='Average Score')

# Show legend
plt.legend()

# Show the chart
plt.tight_layout()
plt.show()

```
